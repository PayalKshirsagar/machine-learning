```python
q1 = df['SALES_PRICE'].quantile(0.25)
```

```python
q3 = df['SALES_PRICE'].quantile(0.75)
```

```python
iqr = q3 - q1
```

```python
upper_limit = q3 + 1.5 * iqr
lower_limit = q1 - 1.5 * iqr
```

```python
upper_limit, lower_limit
```

```python
df[ df['SALES_PRICE'] < lower_limit]
```

```python
df[ df['SALES_PRICE'] > upper_limit]
```

```python
# Removing outliers
new_df = df[(df['SALES_PRICE'] > lower_limit) & (df['SALES_PRICE'] < upper_limit)]
new_df['QS_OVERALL'].dropna()
```

```python
new_df.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['SALES_PRICE'])

plt.subplot(2,2,2)
sns.boxplot(df['SALES_PRICE'])

plt.subplot(2,2,3)
sns.distplot(new_df['SALES_PRICE'])

plt.subplot(2,2,4)
sns.boxplot(new_df['SALES_PRICE'])

plt.show()
```

```python
# capping outliers
new_df_cap = df.copy()

new_df_cap['SALES_PRICE'] = np.where(
    new_df_cap['SALES_PRICE'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['SALES_PRICE'] < lower_limit,
        lower_limit,
        new_df_cap['SALES_PRICE']
    )
)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['SALES_PRICE'])

plt.subplot(2,2,2)
sns.boxplot(df['SALES_PRICE'])

plt.subplot(2,2,3)
sns.distplot(new_df_cap['SALES_PRICE'])

plt.subplot(2,2,4)
sns.boxplot(new_df_cap['SALES_PRICE'])

plt.show()
```

```python
new_df_cap['BUILDTYPE'] = new_df_cap['BUILDTYPE'].replace({
    'Comercial': 'Commercial',
    'Other': 'Others'
})
new_df_cap['PARK_FACIL'] = new_df_cap['PARK_FACIL'].replace({
    'Noo': 'No',
})
new_df_cap = new_df_cap[new_df_cap['QS_OVERALL'].notnull()]
```

```python
new_df_cap['AREA'].value_counts()
```

```python
X = new_df_cap.drop('SALES_PRICE', axis=1)
y = new_df_cap.iloc[:, -3]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
ohe = OneHotEncoder(drop='first',dtype=np.int32)
```

```python
X_train
```

```python
X_train_new = ohe.fit_transform(X_train[['BUILDTYPE','AREA']])
```

```python
X_test_new = ohe.transform(X_test[['BUILDTYPE','AREA']])
```

```python
X_train_new.shape
```

```python
np.hstack(X_train_new)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
X_train
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(drop='first',dtype=np.int32, handle_unknown='ignore'),['BUILDTYPE','AREA']),
    ('tnf2',OrdinalEncoder(categories=[['Yes','No']], handle_unknown='use_encoded_value',
    unknown_value=-1),['PARK_FACIL']),
],remainder='passthrough')
```

```python
X_test
```

```python
transformer.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
model = Pipeline(steps=[
    ('preprocess', transformer),
    ('regressor', LinearRegression())
])
```

```python
X_train.isnull().sum()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

```python
rmse = mean_squared_error(y_test, y_pred)
print("RMSE:", rmse)
```
