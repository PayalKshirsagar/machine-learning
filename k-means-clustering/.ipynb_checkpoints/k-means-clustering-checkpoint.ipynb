```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.columns
```

```python
features = ['enginesize', 'horsepower', 'citympg', 'highwaympg']
X = df[features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd

results = []

for k in [2, 3, 4, 5, 6]:
    kmeans = KMeans(
        n_clusters=k,
        init='random',
        random_state=9001,
        n_init=20
    )
    
    labels = kmeans.fit_predict(X_scaled)
    
    inertia = kmeans.inertia_
    sil_score = silhouette_score(X_scaled, labels)
    
    # cluster counts
    cluster_counts = pd.Series(labels).value_counts().sort_index().to_dict()
    
    results.append({
        'K': k,
        'Inertia': inertia,
        'Silhouette Score': sil_score,
        'Cluster Counts': cluster_counts
    })
```

```python
results_df = pd.DataFrame(results)
results_df
```

```python
inertias = results_df['Inertia'].values
ks = results_df['K'].values

inertia_drops = []

for i in range(1, len(inertias)):
    drop = inertias[i-1] - inertias[i]
    inertia_drops.append(drop)

elbow_df = pd.DataFrame({
    'K_from_to': [f"{ks[i-1]}→{ks[i]}" for i in range(1, len(ks))],
    'Inertia Drop': inertia_drops
})

elbow_df
```

```python
optimal_k = ks[1:][inertia_drops.index(max(inertia_drops))]
optimal_k
```
